```python
import os
original_h5_path = 'drive/MyDrive/github/fund_price_analysis/data/nav'
src_path = 'drive/MyDrive/github/fund_price_analysis/data/index/7days'
fund_info_path = 'drive/MyDrive/github/efficient_selenium/data/fund_info.h5'
index_table_paths = os.listdir(src_path)
```

```python
fund_info_table = pd.read_hdf(fund_info_path)

fund_info_table.columns
```

```python
for col in fund_info_table.columns:
  if fund_info_table[col].nunique() < 100:
    print(col, ':', fund_info_table[col].unique())
```

```python
import pandas as pd
import tqdm
import concurrent.futures
# TODO: last_sharpxxx add where to find the last date info
def get_last_sharp_ratio(path):
  try:
    return path, pd.read_hdf(os.path.join(src_path, path)).sharp_ratio.values[-1]
  except:
    return path, None
def get_std(path):
  try:
    return path, pd.read_hdf(os.path.join(src_path, path)).earning_rate_std.values[-1]
  except:
    return path, None
def get_mean(path):
  try:
    return path, pd.read_hdf(os.path.join(src_path, path)).earning_rate_mean.values[-1]
  except:
    return path, None

def get_last_sortino_ratio(path):
  try:
    return path, pd.read_hdf(os.path.join(src_path, path)).sortino_ratio.values[-1]
  except:
    return path, None

def get_plausible_length(path):
  try:
    return path, len(pd.read_hdf(os.path.join(src_path, path)))
  except:
    return path, None

def get_plausible_ratio(path):
  try:
    ratio = len(pd.read_hdf(os.path.join(src_path, path))) / len(pd.read_hdf(os.path.join(original_h5_path, path)))
    return path, ratio
  except:
    return path, None

def get_history_length(path):
  try:
    return path, len(pd.read_hdf(os.path.join(original_h5_path, path)))
  except:
    return path, None
```

```python
def get_fund_and_company_name(path):
  isin_str = path.split('.h5')[0]
  fund_info = pd.read_hdf(fund_info_path, where=f'ISIN="{isin_str}"')
  return path, fund_info['基金名稱'].values[0], fund_info['基金管理公司'].values[0]
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:
  fund_and_company_names = list(filter(
      lambda x: (x[0] is not None) and (x[1] is not None), 
      tqdm.tqdm(
          executor.map(get_fund_and_company_name, index_table_paths), 
          total=len(index_table_paths))
      ))
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:
  earning_means = list(filter(
      lambda x: x[1] is not None, 
      tqdm.tqdm(
          executor.map(get_mean, index_table_paths), 
          total=len(index_table_paths))
      ))
  earning_stds = list(filter(
      lambda x: x[1] is not None, 
      tqdm.tqdm(
          executor.map(get_std, index_table_paths), 
          total=len(index_table_paths))
      ))
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:
  sharp_ratios = list(filter(
      lambda x: x[1] is not None, 
      tqdm.tqdm(
          executor.map(get_last_sharp_ratio, index_table_paths), 
          total=len(index_table_paths))
      ))
  sortino_ratios = list(filter(
      lambda x: x[1] is not None, 
      tqdm.tqdm(
          executor.map(get_last_sortino_ratio, index_table_paths), 
          total=len(index_table_paths))
      ))
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:
  plausible_lengths = list(filter(
      lambda x: x[1] is not None, 
      tqdm.tqdm(
          executor.map(get_plausible_length, index_table_paths), 
          total=len(index_table_paths))
      ))
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:
  plausible_ratios = list(filter(
      lambda x: x[1] is not None, 
      tqdm.tqdm(
          executor.map(get_plausible_ratio, index_table_paths), 
          total=len(index_table_paths))
      ))
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:
  history_lengths = list(filter(
      lambda x: x[1] is not None, 
      tqdm.tqdm(
          executor.map(get_history_length, index_table_paths), 
          total=len(index_table_paths))
      ))
```
